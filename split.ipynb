```python
import os 
import numpy as np

feature_path= '/space/gzanardini/tuh_background/'
outpath = '/space/gzanardini/tuh_background/split/'
if not os.path.exists(outpath):
    os.makedirs(outpath)

file_list = [ file for file in os.listdir(feature_path) if file.endswith('.npy')]
```

```python
print(file_list)
```

```python
for file in file_list:
   if file.endswith('.npy'):
      data=np.load(os.path.join(feature_path, file))

      print (data.shape)
      data_mean=data[:,0,:]
      data_median=data[:,1,:]
      data_std=data[:,2,:]
      data_skew=data[:,3,:]
      data_kurtosis=data[:,4,:]
      
      length = file.split('_')[2]
      np.save(os.path.join(outpath, file.replace(length, length.replace('.npy', '_mean.npy'))), data_mean)
      np.save(os.path.join(outpath, file.replace(length, length.replace('.npy', '_median.npy'))), data_median)
      np.save(os.path.join(outpath, file.replace(length, length.replace('.npy', '_std.npy'))), data_std)
      np.save(os.path.join(outpath, file.replace(length, length.replace('.npy', '_skew.npy'))), data_skew)
      np.save(os.path.join(outpath, file.replace(length, length.replace('.npy', '_kurt.npy'))), data_kurtosis)
```
