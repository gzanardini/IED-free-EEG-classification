```python
import os 
import numpy as np

feature_path= '/space/gzanardini/tuh_epoched/'
outpath = '/space/gzanardini/tuh_epoched/split/'

file_list = [ file for file in os.listdir(feature_path) if file.endswith('.npy')]
```

```python
print(file_list)
```

```python
for file in file_list:
   if file.split('_')[2]=='10s.npy':
        data=np.load(os.path.join(feature_path, file))

        print (data.shape)
        data_mean=data[:,0,:]
        data_median=data[:,1,:]
        data_std=data[:,2,:]
        data_skew=data[:,3,:]
        data_kurtosis=data[:,4,:]
    
        np.save(os.path.join(outpath, file.replace('10s.npy', '10s_mean.npy')), data_mean)
        np.save(os.path.join(outpath, file.replace('10s.npy', '10s_median.npy')), data_median)
        np.save(os.path.join(outpath, file.replace('10s.npy', '10s_std.npy')), data_std)
        np.save(os.path.join(outpath, file.replace('10s.npy', '10s_skew.npy')), data_skew)
        np.save(os.path.join(outpath, file.replace('10s.npy', '10s_kurt.npy')), data_kurtosis)
```
