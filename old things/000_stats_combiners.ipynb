```python
# Perform Mann-Whitney U test
def mannwhitneyu(data1, data2):
    u_stat, p_val = stats.mannwhitneyu(data1, data2)
    return u_stat, p_val
```

## Features - Statistical combiners

```python
subjects_list=description['subject'].to_numpy()
subjects=[]
sessions=[]
epilepsy_labels=[]
indexes=[]

for i in range(len(description)):
    subject = description.iloc[i]['subject']
    session = description.iloc[i]['session']
    epilepsy_status = description.iloc[i]['epilepsy']
    idxs = np.where((subjects_list == subject) & (description['session'] == session) & (labels == epilepsy_status) & (description['frequency'] != 23))[0]

    if not any(np.array_equal(idxs, x) for x in indexes):
        indexes.append(idxs)
        subjects.append(subject)
        sessions.append(session)
        epilepsy_labels.append(epilepsy_status)

subjects=np.array(subjects)
sessions=np.array(sessions)
epilepsy_labels=np.array(epilepsy_labels)

aggregation_df = pd.DataFrame([subjects, sessions, epilepsy_labels, indexes]).T
aggregation_df.columns = ['subject', 'session', 'epilepsy', 'indexes']

aggregation_df.to_csv('data/feats/aggregation_df.csv', index=False)
```

```python
for i in range(len(aggregation_df)):
    if len(aggregation_df.loc[i]['indexes']) > 11:
        if len(aggregation_df.loc[i]['indexes']) ==12:
            aggregation_df.loc[i]['indexes'] = aggregation_df.loc[i]['indexes'][:11]
        else:
            line=aggregation_df.loc[i]
            temp_idxs = aggregation_df.loc[i]['indexes']
            top_idxs = temp_idxs[:11]
            bottom_idxs = temp_idxs[11:]

            print(top_idxs),print(len(top_idxs))
            print(bottom_idxs),print(len(bottom_idxs))

            aggregation_df.loc[i]['indexes'] = top_idxs
            aggregation_df = aggregation_df.append({'subject': line['subject'], 'session': line['session'], 'epilepsy': line['epilepsy'], 'indexes': bottom_idxs}, ignore_index=True)

aggregation_df.to_csv('data/feats/aggregation_df.csv', index=False)
```

```python
indexes=aggregation_df['indexes'].to_numpy()

for montage, montage_name in zip(montage_sets, montages):
    if montage_name == 'CAR':
        data_concat = data_concat_car
    elif montage_name == 'Cz':
        data_concat = data_concat_cz
    elif montage_name == 'BipolarDB':
        data_concat = data_concat_bipolar
    elif montage_name == 'Laplacian':
        data_concat = data_concat_laplacian
    
    data_combined = []

    for index_list in indexes:
        data_mean = np.mean(data_concat[index_list], axis=0).reshape(1,-1)
        data_std = np.std(data_concat[index_list], axis=0).reshape(1,-1)
        data_median = np.median(data_concat[index_list], axis=0).reshape(1,-1)
        data_skewness = skew(data_concat[index_list], axis=0).reshape(1,-1)
        data_kurtosis = kurtosis(data_concat[index_list], axis=0).reshape(1,-1)
        data_concatenated = np.concatenate([data_mean, data_std, data_median, data_skewness, data_kurtosis], axis=1)
        data_combined.append(data_concatenated)

    data_combined = np.array(data_combined).squeeze()

    print(f'{montage_name} shape: {np.array(data_combined).shape}')
    np.save(f'data/feats/aggregated_features_{montage_name}.npy', np.array(data_combined))
```

```python
car_aggregated = np.load('data/feats/aggregated_features_CAR.npy')
cz_aggregated = np.load('data/feats/aggregated_features_Cz.npy')
bipolar_aggregated = np.load('data/feats/aggregated_features_BipolarDB.npy')
laplacian_aggregated = np.load('data/feats/aggregated_features_Laplacian.npy')

labels=aggregation_df['epilepsy'].to_numpy()

car_mannwhitneyu_results=[]

for row in car_aggregated.T:
    u_stat, p_val = mannwhitneyu(row[labels==0], row[labels==1])
    car_mannwhitneyu_results.append(p_val)

cz_mannwhitneyu_results=[]

for row in cz_aggregated.T:
    u_stat, p_val = mannwhitneyu(row[labels==0], row[labels==1])
    cz_mannwhitneyu_results.append(p_val)

bipolar_mannwhitneyu_results=[]
for row in bipolar_aggregated.T:
    u_stat, p_val = mannwhitneyu(row[labels==0], row[labels==1])
    bipolar_mannwhitneyu_results.append(p_val)

laplacian_mannwhitneyu_results=[]
for row in laplacian_aggregated.T:
    u_stat, p_val = mannwhitneyu(row[labels==0], row[labels==1])
    laplacian_mannwhitneyu_results.append(p_val)
```

```python
#print the number of features that are significant

significance_threshold=0.01

print(f'CAR: {np.sum(np.array(car_mannwhitneyu_results) < significance_threshold)}')
print(f'Cz: {np.sum(np.array(cz_mannwhitneyu_results) < significance_threshold)}')
print(f'Bipolar: {np.sum(np.array(bipolar_mannwhitneyu_results) < significance_threshold)}')
print(f'Laplacian: {np.sum(np.array(laplacian_mannwhitneyu_results) < significance_threshold)}')

# Save the significant features
np.save(f'aggre_feats/car_mannwhitneyu_results_{significance_threshold}.npy', car_mannwhitneyu_results)
np.save(f'aggre_feats/cz_mannwhitneyu_results_{significance_threshold}.npy', cz_mannwhitneyu_results)
np.save(f'aggre_feats/bipolar_mannwhitneyu_results_{significance_threshold}.npy', bipolar_mannwhitneyu_results)
np.save(f'aggre_feats/laplacian_mannwhitneyu_results_{significance_threshold}.npy', laplacian_mannwhitneyu_results)
```
