```python
import os
import numpy as np

feature_dir='/space/gzanardini/tuh_epoched/split'

file_list=[file for file in os.listdir(feature_dir) if file.endswith('.npy')]
file_list.sort()
```

```python
# the feature files come in the format type_features_montage_sLength_combiner.npy
# rename them to type_montage_Lengths_combiner.

for file in file_list:
    print(file)
    parts = file.split('_')
    print(parts)
    parts.remove('features')  # remove 'features'
    print(parts)
    parts[2] = parts[2].replace('s', '')  # remove 's' from Length
    print(parts)
    #append s to parts[3]  # ensure Length has 's' at the end
    parts[2] += 's'  # ensure Length has 's' at the end
    print(parts)

    new_name = '_'.join(parts)
    print(new_name)

    os.rename(os.path.join(feature_dir, file), os.path.join(feature_dir, new_name))
```
