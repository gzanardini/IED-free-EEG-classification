```python
import numpy as np 
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Load data
description=pd.read_csv('data/photostim_steps.csv')
labels=description['epilepsy'].to_numpy()
subjects=description['subject'].to_numpy()
unique_subjects=np.unique(description['subject'])

cwt_car =     np.load(f'data/featsv2/cwt_features_CAR.npy')
cwt_cz =      np.load(f'data/featsv2/cwt_features_Cz.npy')
cwt_bipolar = np.load(f'data/featsv2/cwt_features_BipolarDB.npy')
cwt_laplacian = np.load(f'data/featsv2/cwt_features_Laplacian.npy')

dwt_car =     np.load(f'data/featsv2/dwt_features_CAR.npy')
dwt_cz =      np.load(f'data/featsv2/dwt_features_Cz.npy')
dwt_bipolar = np.load(f'data/featsv2/dwt_features_BipolarDB.npy')
dwt_laplacian = np.load(f'data/featsv2/dwt_features_Laplacian.npy')

gplv_car =     np.load(f'data/featsv2/gplv_features_CAR.npy')
gplv_cz =      np.load(f'data/featsv2/gplv_features_Cz.npy')
gplv_bipolar = np.load(f'data/featsv2/gplv_features_BipolarDB.npy')
gplv_laplacian = np.load(f'data/featsv2/gplv_features_Laplacian.npy')

mst_car =     np.load(f'data/featsv2/mst_features_CAR.npy')
mst_cz =      np.load(f'data/featsv2/mst_features_Cz.npy')
mst_bipolar = np.load(f'data/featsv2/mst_features_BipolarDB.npy')
mst_laplacian = np.load(f'data/featsv2/mst_features_Laplacian.npy')

spectral_car =     np.load(f'data/featsv2/spectral_features_CAR.npy')
spectral_cz =      np.load(f'data/featsv2/spectral_features_Cz.npy')
spectral_bipolar = np.load(f'data/featsv2/spectral_features_BipolarDB.npy')
spectral_laplacian = np.load(f'data/featsv2/spectral_features_Laplacian.npy')

sst_car =     np.load(f'data/featsv2/sst_features_CAR.npy')
sst_cz =      np.load(f'data/featsv2/sst_features_Cz.npy')
sst_bipolar = np.load(f'data/featsv2/sst_features_BipolarDB.npy')
sst_laplacian = np.load(f'data/featsv2/sst_features_Laplacian.npy')

utm_car =     np.load(f'data/featsv2/utm_features_CAR.npy')
utm_cz =      np.load(f'data/featsv2/utm_features_Cz.npy')
utm_bipolar = np.load(f'data/featsv2/utm_features_BipolarDB.npy')
utm_laplacian = np.load(f'data/featsv2/utm_features_Laplacian.npy')

plv_car =     np.load(f'data/featsv2/plv_features_CAR.npy')
plv_cz =      np.load(f'data/featsv2/plv_features_Cz.npy')
plv_bipolar = np.load(f'data/featsv2/plv_features_BipolarDB.npy')
plv_laplacian = np.load(f'data/featsv2/plv_features_Laplacian.npy')
```

```python
car = [cwt_car, dwt_car, gplv_car, mst_car, spectral_car, sst_car, utm_car, plv_car]
cz = [cwt_cz, dwt_cz, gplv_cz, mst_cz, spectral_cz, sst_cz, utm_cz, plv_cz]
bipolar = [cwt_bipolar, dwt_bipolar, gplv_bipolar, mst_bipolar, spectral_bipolar, sst_bipolar, utm_bipolar, plv_bipolar]
laplacian = [cwt_laplacian, dwt_laplacian, gplv_laplacian, mst_laplacian, spectral_laplacian, sst_laplacian, utm_laplacian, plv_laplacian]

feat_names = ['cwt', 'dwt', 'gplv', 'mst', 'spectral', 'sst', 'utm', 'plv']

montages=['CAR', 'Cz', 'BipolarDB', 'Laplacian']
montage_sets =[car, cz, bipolar, laplacian]

# Perform Mann-Whitney U test
def mannwhitneyu(data1, data2):
    u_stat, p_val = stats.mannwhitneyu(data1, data2)
    return u_stat, p_val
```

```python
def custom_scale(x, pos):
    if x < 0.2:
        return f'{x:.1e}'
    else:
        return f'{x:.2f}'
```

```python
significance_threshold=0.001
```

```python
for montage_set, montage_name in zip(montage_sets, montages):

    for (set_car, feat_name) in zip(montage_set, feat_names):
        mannwhitneyu_results = []

        print(f'Performing Mann-Whitney U test for {feat_name} features ({montage_name})...')

        for row in set_car.T:
            u_stat, p_val = mannwhitneyu(row[labels==0], row[labels==1])
            mannwhitneyu_results.append(p_val)

        mannwhitneyu_results = np.array(mannwhitneyu_results)

        # Plot p-values
        plt.figure(figsize=(20, 5))
        #plot a horizontal line at p=significance_threshold

        plt.axhline(significance_threshold, color='red', linestyle='--')
        plt.plot(mannwhitneyu_results, 'o')

        ax = plt.gca()


        # color the significant p-values
        for i, p in enumerate(mannwhitneyu_results):
            if p < significance_threshold:
                ax.plot(i, p, 'o', color='red')

        ax.set_yscale('log')
        ax.set_xlabel('Feature index')
        ax.set_ylabel('p-value')
        ax.yaxis.set_major_formatter(plt.FuncFormatter(custom_scale))
        ax.title.set_text('Mann-Whitney U test p-values for ' + feat_name + ' features ('+montage_name+')')
        print(f'Number of significant p-values: {np.sum(mannwhitneyu_results < significance_threshold)} out of {len(mannwhitneyu_results)}')

        idxs=np.where(mannwhitneyu_results < significance_threshold)[0]

        print(f'Significant p-values indexes ({montage_name}): {idxs}')

        np.save(f'significance/{feat_name}_features_{montage_name}.npy', idxs)

        plt.show()
```
