```python
import mne 
import numpy as numpy
import pickle as pkl
import pandas as pd
import os
import matplotlib as mpl
# mpl.use('Qt5Agg')
import matplotlib.pyplot as plt
```

```python
cwd = os.getcwd()
print(cwd)

samples_relpath = '/temp_data/stim_samples.pkl'
df_relpath = '/temp_data/stim_df.csv'

stim_samples = pkl.load(open(cwd + samples_relpath, 'rb'))
stim_df = pd.read_csv(cwd + df_relpath)
```

```python
stim_samples[0]['PHOTIC PH'][0].shape
```

```python
for sample in stim_samples:
    data=sample.get_data()

    plt.figure(figsize=(10, 5))
    plt.subplot(211)
    plt.plot(data[19])
    plt.title('Stimulus')
    plt.subplot(212)
    plt.title('Channel O2')
    plt.plot(data[18])
    plt.show()
    plt.pause(1)
```

```python
steps_relpath = '/data/photostim_steps.npy'
steps_df_relpath='/data/photostim_steps.csv'

steps = numpy.load(cwd + steps_relpath)
steps_df = pd.read_csv(cwd + steps_df_relpath)
```

```python
steps_df
```

```python
from matplotlib.pylab import f

with plt.ioff():
    for i,step in enumerate(steps):
        plt.figure(figsize=(20, 5))
        plt.subplot(211)
        plt.title('Stimulus')
        plt.plot(step[19])
        plt.subplot(212)
        plt.title('O2 Channel')
        plt.plot(step[18])
        plt.suptitle(f'Step {i} - Subject {steps_df.iloc[i]["subject"]}, Session{steps_df.iloc[i]["session"]} - Frequency {steps_df.iloc[i]["frequency"]} Hz')
        plt.show()

    
```
